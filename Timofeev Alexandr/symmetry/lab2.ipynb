```python
import itertools
```

```python
def NonEquivalentBitstrings(n, k):

    source = '0'*(n-k)

    unique = []

    # исходная строка содержит только нули
    # из нее можно получить любую двоичную последовательность с n-k нулями
    # для этого из n-k+1 мест выберем k с повторением и вставим в эти места единицы
    # вставка единицы смещает индексы, но комбинации выдаются в порядке возрастания индексов,
    # поэтому можно завести счетчик количества вставленных единиц и прибавлять к каждому 
    # исходному выбранному индексу кол-во единиц, вставленных до него (они все слева)
    for comb in itertools.combinations_with_replacement(range(n-k+1), k): 
        s = source
        i = 0 # счетчик вставленных единиц
        for c in comb:
            s = s[:c+i] + '1' + s[c+i:]
            i += 1

        # если ни последовательность, ни одна из
        # последовательностей, эквивалентных ей 
        # еще не найдены, добавляем
        for _ in range(n):
            if s in unique:
                break
            s = s[-1] + s[:-1]
        else:
            unique.append(s)
    
    return unique
```

```python
def findSymmetryAxesEven(bitstring):
    n = len(bitstring)
    axes = set()
    
    for _ in range(n//2):
        if bitstring[n//2-1::-1] == bitstring[n//2:]: # не проходящие через вершины
            axes.add(bitstring[:n//2] + '|' + bitstring[n//2:])
            
        if bitstring[n//2-1:0:-1] == bitstring[n//2+1:]: # проходящие через вершны
            axes.add('(' + bitstring[0] + ')' + bitstring[1:n//2] + '(' + bitstring[n//2] + ')' + bitstring[n//2+1:])
        
        bitstring = bitstring[-1] + bitstring[:-1]
        
    return axes

def findSymmetryAxesOdd(bitstring):
    n = len(bitstring)
    axes = set()
    for _ in range(n):
        if bitstring[n//2-1::-1] == bitstring[n//2+1:] : # проходящие через одну вершину
            axes.add(bitstring[:n//2] + '(' + bitstring[n//2] + ')' + bitstring[n//2+1:])
    
        bitstring = bitstring[-1] + bitstring[:-1]
    
    return axes
```

```python
def findSymmetryAxes(bitstrings):
    if len(bitstrings[0]) % 2 == 0:
        func = findSymmetryAxesEven
    else:
        func = findSymmetryAxesOdd
    
    syms = {key:set() for key in bitstrings}

    for string in bitstrings:
        for s in func(string):
            syms[string].add(s)
    
    return syms
```

```python
for n in range(3, 10): # вывод
    for k in range(1, n):
        print(f'n={n}, k={k}')
        
        axes = findSymmetryAxes(NonEquivalentBitstrings(n, k))
        # последовательности из одного класса
        # эквивалентности выводятся в одной строке
        for key in axes.keys(): 
            syms = list(axes[key])
            if (len(syms) > 0):
                print(*syms)
```
